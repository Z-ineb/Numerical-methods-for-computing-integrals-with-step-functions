**Midpoint method**

*f(x) is continuous on  [a,b]
<br>
 n a positive integer and  $Δx$=$ b−a \over n$. 
<br>
If  [a,b] is divided into  n subintervals, each of length  $Δx$, and  $m_i$ is the midpoint of the  ith subinterval
<br>
the integral of f(x) between a,b with the midpoint rule is
<br>
 $Δx\sum_{i=1}^n m_i$*

```python
#difine the midpoint method
def midpoint(f, a, b, n):
  h = float(b-a)/n # it's the width of each subinterval 
  result =0
  for i in range(n):
   result += f((a + h/2.0) + i*h) # represent f(xi)
   result *= h
  return result
```

```python
# example with f(x)=x2 +1 
def f(x):
    return x**2+1

a = -1
b = 2
n = 3

print("with midpoint rule = ", midpoint(f, a, b, n))
```

```python
import warnings
warnings.filterwarnings("ignore")

def integArea(eq, upperLim, bottomLim, breakUp):
    f = eq
    a = upperLim; b = bottomLim; N = breakUp
    x = np.linspace(a,b,N+1)
    X = np.linspace(a,b,N*N+1)
    do = np.linspace(a,b,1000)
    Y = f(X)

    plt.figure(figsize=(15,5))

    plt.subplot(1,3,3)
    plt.plot(X,Y,'b')
    x_mid = (x[:-1] + x[1:])/2 # Midpoints
    y_mid = f(x_mid)
    plt.plot(x_mid,y_mid,'r.',markersize=10)
    plt.bar(x_mid,y_mid,width=(b-a)/N,color="#0f0f0f50")
    plt.title('Midpoint Rule, N = {}'.format(N))

integArea(f, a, b, n)
```

```python
#exaple with the fuction f(x)=sin(x)
def f1(x):
    return np.sin(x)

a1 = 0
b1 = np.pi
n1 = 3

integArea(f1,a1,b1,n1)
print("with midpoint rule = ", midpoint(f1, a1, b1, n1))
```

**Trapezoidal method**

*f(x) is continuous on  [a,b]
<br>
 n a positive integer and $Δx$= $(b−a) \over n$. 
<br>
If  [a,b] is divided into  n subintervals, each of length  $Δx$ 
<br>
the integral of f(x) between a,b with the Trapezoidal rule is
<br>
<br>
 $Δx \over 2$ $[f(x_0)+2f(x_1)+2f(x_2)+.........+2f(x_{n-1})+f(x_n)]$*

```python
# Trapezoidal Rule
def trapezoidalRule(f,a,b,n):
  h = float(b-a)/n
  f_sum = 0
  for i in range(1, n, 1):
    x = a + i * h
    f_sum = f_sum + f(x)
  return format((h * (0.5 * f(a) + f_sum + 0.5 * f(b))), '.16f')
```

```python
print("with Trapezoidal rule = ", trapezoidalRule(f, a, b, n))
```

**Simpson Method**

*f(x) is continuous on  [a,b]
<br>
 n a positive integer and $Δx$= $(b−a) \over n$. 
<br>
If  [a,b] is divided into  n subintervals, each of length  $Δx$ 
<br>
the integral of f(x) between a,b with the simpson's rule is
<br>
<br>
 $Δx \over 3$ $[(f(x_0)+4f(x_1)+2f(x_2)+4f(x_3)+2f(x_5).........+2f(x_{n-2})+4f(x_{n-1})+f(x_n)]$*

```python
# Simpson Rule
def simpsonRule(F,a,b,n):
  h = float(b-a) / n
  result = F(a) + F(b)
  for i in range(1, n, 1):
    if(i % 2 == 0):
      result = result + (2*(f(a+i*h)))
    else :
      result = result + (4*(f(a+i*h)))
  result *= (h/3)
  return format(result, '.16f')
```

```python
print("with simpson rule = ", simpsonRule(f, a, b, n))
```
